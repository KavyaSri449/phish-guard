#Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = df.select_dtypes(include=['float64', 'int64']).columns
# Calculate the number of rows and columns for the subplot grid
num_rows = (len(num_cols) + 5) // 6  # Calculate rows, ensuring enough space
# +5 to round up for the leftover plots, if any

plt.figure(figsize=(20, 15))
for i, col in enumerate(num_cols, 1):
    plt.subplot(num_rows, 6, i)  # Dynamically adjust grid size
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

#x and y split

```python
x = df.iloc[:, :-1]
x.head()
```

```python
y = df.Result
print("y:",y.head())
```

#Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_tr, x_t, y_tr, y_t = train_test_split(x, y, test_size=0.2, random_state=0)
```

#Model Building

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_tr,y_tr)
```

```python
y_plr=lr.predict(x_t)
y_plr1=lr.predict(x_tr)
```

```python
from sklearn.metrics import accuracy_score
test_acc=accuracy_score(y_t,y_plr)
train_acc=accuracy_score(y_tr,y_plr1)
print("Accuracy score for testing data: ",test_acc)
print("Accuracy score for training data: ",train_acc)
```

#Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_t, y_plr)
print("Mean Squared Error:", mse)
rmse = mean_squared_error(y_t, y_plr)
print("Root Mean Squared Error:", rmse)
r2 = r2_score(y_t, y_plr)
print("R-squared:", r2)
```
